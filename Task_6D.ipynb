#### Query 2

```python
# Extracting unique combinations of 'type' and 'engine' columns from the 'planes' DataFrame.
# Reset the index for better presentation.
task2_my = ( planes.loc[:,['type', 'engine']].drop_duplicates().reset_index(drop=True))
task2_my.head(5)
```

#### Query 3

```python
task3_sql = pd.read_sql_query("""SELECT COUNT(*), engine FROM planes GROUP BY engine""", conn)
task3_sql.head(3)
```

```python
# Group the DataFrame 'planes' based on the 'engine' column and calculate the count of each engine type.
# Reset the index and rename the columns for clarity, resulting in the 'task3_my' DataFrame.
task3_my = ( planes.loc[:,['engine']].groupby('engine')['engine'].size().reset_index(name='COUNT(*)').loc[:,["COUNT(*)","engine"]])
task3_my.head(5)
```

#### Query 4

```python
task4_sql = pd.read_sql_query("""SELECT COUNT(*), engine, type FROM planes GROUP BY engine, type""", conn)
task4_sql.head(3)
```

```python
# Group the DataFrame 'planes' based on 'engine' and 'type' columns.
# Calculate the count of occurrences for each combination of 'engine' and 'type'.
# Reset the index and rename columns to create the 'task4_my' DataFrame.

task4_my = (planes.loc[:,['engine','type']].groupby(['engine','type'])['engine'].size().reset_index(name='COUNT(*)').loc[:,["COUNT(*)","engine","type"]])
task4_my.head(5)
```

```python
# Query 5
task5_sql = pd.read_sql_query("""SELECT MIN(year), AVG(year), MAX(year), engine, manufacturer FROM planes GROUP BY engine, manufacturer""", conn)
task5_sql.head()
```

```python
# Group the DataFrame 'planes' based on 'engine' and 'manufacturer' columns.
# Calculate the minimum, mean, and maximum values of the 'year' for each combination.
# Rename the aggregated columns and reset the index to create the 'task5_my' DataFrame.
task5_my = (planes.loc[:,['year','engine','manufacturer']].groupby(['engine','manufacturer'])['year'].aggregate([np.min,np.mean,np.max]).rename(columns={"min":"MIN(year)", "mean":"AVG(year)", "max":"MAX(year)"}).reset_index().loc[:,["MIN(year)","AVG(year)","MAX(year)","engine", "manufacturer"]])
task5_my.head(5)
```

```python
# Query 6
task6_sql = pd.read_sql_query("""SELECT * FROM planes WHERE speed IS NOT NULL""", conn)
task6_sql.head()
```

```python
# Remove rows from the 'planes' DataFrame where the 'speed' column is null.
# Reset the index to create the 'task6_my' DataFrame.
task6_my = (planes.loc[~planes['speed'].isnull(),:].reset_index(drop=True))
task6_my.head(5)
```

```python
# Query 7
task7_sql = pd.read_sql_query("""SELECT tailnum FROM planes WHERE seats BETWEEN 150 AND 210 AND year >= 2011""", conn)
task7_sql.head()
```

```python
# Filter the 'planes' DataFrame to include rows where the 'seats' are between 150 and 210 (inclusive) and the 
# 'year' is greater than or equal to 2011. Extract the 'tailnum' column and reset the index.
task7_my = (planes.loc[(planes.seats>=150) & (planes.seats<=210) & (planes.year>=2011), ['tailnum']].reset_index(drop=True))
task7_my.head(5)
```

```python
# Query 8
task8_sql = pd.read_sql_query("""SELECT tailnum, manufacturer, seats FROM planes WHERE manufacturer IN ("BOEING", "AIRBUS", "EMBRAER") AND seats>390""", conn)
task8_sql.head()
```

```python
# Filter the 'planes' DataFrame to include rows where the 'manufacturer' is either "BOEING", "AIRBUS", or "EMBRAER",
# and the 'seats' are greater than 390. Extract the 'tailnum', 'manufacturer', and 'seats' columns and reset the index.
task8_my = (planes.loc[(planes.manufacturer.isin(["BOEING", "AIRBUS", "EMBRAER"])) & (planes.seats>390),['tailnum', 'manufacturer', 'seats']].reset_index(drop=True))
task8_my.head(5)
```

```python
# Query 9
task9_sql = pd.read_sql_query("""SELECT DISTINCT year, seats FROM planes WHERE year >= 2012 ORDER BY year ASC, seats DESC""", conn)
task9_sql.head()
```

```python
# Filter the 'planes' DataFrame to include rows where the 'year' is greater than or equal to 2012.
# Extract the 'year' and 'seats' columns, remove duplicates, sort by 'year' in ascending order and 'seats' in descending order,
# and reset the index.
task9_my = (planes.loc[(planes.year>=2012),['year', 'seats']].drop_duplicates().sort_values(['year','seats'], ascending=[True, False]).reset_index(drop=True))
task9_my.head(5)
```

```python
# Query 10
task10_sql = pd.read_sql_query("""SELECT DISTINCT year, seats FROM planes WHERE year >= 2012 ORDER BY seats DESC, year ASC""", conn)
task10_sql.head()
```

```python
# Filter the 'planes' DataFrame to include rows where the 'year' is greater than or equal to 2012.
# Extract the 'year' and 'seats' columns, remove duplicates, sort by 'seats' in descending order and 'year' in ascending order,
# and reset the index.

task10_my = (planes.loc[(planes.year>=2012),['year', 'seats']].drop_duplicates().sort_values(['seats','year'], ascending=[False, True]).reset_index(drop=True))
task10_my.head(5)
```

```python
# Query 11
task11_sql = pd.read_sql_query("""SELECT manufacturer, COUNT(*) FROM planes WHERE seats > 200 GROUP BY manufacturer""", conn)
task11_sql.head()
```

```python
# Filter the 'planes' DataFrame to include rows where 'seats' is greater than 200.
# Group by 'manufacturer' and count the occurrences, reset the index, and rename the count column.

task11_my = (planes.loc[(planes.seats>200),['manufacturer']].groupby('manufacturer')['manufacturer'].size().reset_index(name='COUNT(*)'))
task11_my
```

```python
# Query 12
task12_sql = pd.read_sql_query("""SELECT manufacturer, COUNT(*) FROM planes GROUP BY manufacturer HAVING COUNT(*) > 10""", conn)
task12_sql.head()
```

```python
# Group the 'planes' DataFrame by 'manufacturer' and count the occurrences, reset the index, and rename the count column.
# Filter the result to include only manufacturers with more than 10 occurrences.
# Sort the result by 'manufacturer' in ascending order and reset the index.

task12_my = (planes.loc[:,['manufacturer']].groupby('manufacturer')['manufacturer'].size().reset_index(name="COUNT(*)"))
task12_my = (task12_my.loc[task12_my['COUNT(*)']>10, :].sort_values("manufacturer", ascending=True).reset_index(drop=True))
task12_my.head(5)
```

```python
# Query 13
task13_sql = pd.read_sql_query("""SELECT manufacturer, COUNT(*) FROM planes WHERE seats > 200 GROUP BY manufacturer HAVING COUNT(*) > 10""", conn)
task13_sql.head()
```

```python
# Filter the 'planes' DataFrame to include only rows where 'seats' are greater than 200.
# Group the filtered DataFrame by 'manufacturer' and count the occurrences, reset the index, and rename the count column.
# Filter the result to include only manufacturers with more than 10 occurrences.
# Sort the result by 'manufacturer' in ascending order and reset the index.

task13_my = (planes.loc[(planes.seats>200),['manufacturer']].groupby('manufacturer')['manufacturer'].size().reset_index(name="COUNT(*)"))
task13_my = (task13_my.loc[task13_my['COUNT(*)']>10, :].sort_values("manufacturer", ascending=True).reset_index(drop=True))
task13_my.head(5)
```

```python
# Query 14
task14_sql = pd.read_sql_query("""SELECT manufacturer, COUNT(*) AS howmany FROM planes GROUP BY manufacturer ORDER BY howmany DESC LIMIT 10""", conn)
task14_sql.head()
```

```python
# Group the 'planes' DataFrame by 'manufacturer' and count the occurrences, rename the count column to 'howmany', reset the index.
# Sort the result by 'howmany' in descending order and reset the index.
# Select the top 10 rows from the sorted result.

task14_my = (planes.loc[:,['manufacturer']].groupby('manufacturer')['manufacturer'].size().rename("howmany").reset_index().sort_values("howmany", ascending=False).reset_index(drop=True).head(10))
task14_my.head(5)
```

```python
# Query 15
task15_sql = pd.read_sql_query("""SELECT flights.*,planes.year AS plane_year,planes.speed AS plane_speed,planes.seats AS plane_seats FROM flights LEFT JOIN planes ON flights.tailnum=planes.tailnum""", conn)
task15_sql.head()
```

```python
# Merge the 'flights' and 'planes' DataFrames based on the 'tailnum' column.
# Select specific columns ('tailnum', 'year', 'speed', 'seats') from the 'planes' DataFrame and rename them.
# Merge the selected columns back into the 'flights' DataFrame using the 'tailnum' column as the key.

task15_my = pd.merge(flights,planes.loc[:,["tailnum","year","speed","seats"]].rename(columns={"year":"plane_year", "speed":"plane_speed", "seats":"plane_seats"}),on="tailnum",how="left")
task15_my.head(5)
```

```python
# Query 16
task16_sql = pd.read_sql_query("""SELECT planes.*, airlines.*FROM (SELECT DISTINCT carrier, tailnum FROM flights) AS cartail INNER JOIN planes ON cartail.tailnum=planes.tailnum INNER JOIN airlines ON cartail.carrier=airlines.carrier""", conn)
task16_sql.head()
```

```python
# Create a list of column names from the 'planes' DataFrame and add the column 'carrier' to it.
cols = planes.columns.values.tolist() + ["carrier"]

# Select unique combinations of 'carrier' and 'tailnum' columns from the 'flights' DataFrame and reset the index.
cartail = (flights.loc[:,["carrier","tailnum"]].drop_duplicates().reset_index(drop=True))

# Merge the result with the 'planes' DataFrame on the 'tailnum' column using inner join.
# Select the columns from the merged DataFrame based on the combined list of column names.
# Merge the result with the 'airlines' DataFrame on the 'carrier' column using inner join.
# Sort the merged DataFrame based on 'tailnum' and 'carrier' columns in ascending order.
# Reset the index of the final DataFrame.
task16_my = (pd.merge(cartail,planes,on='tailnum',how='inner')
 .reset_index(drop=True).loc[:,cols].merge(airlines,on='carrier',how='inner')
 .sort_values(['tailnum','carrier'], ascending=[True, True])
 .reset_index(drop=True)
)
task16_my.head(3)
```

```python
# Query 17
task17_sql = pd.read_sql_query("""SELECT flights2.*,atemp,ahumid 
FROM (SELECT * FROM flights WHERE origin='EWR') 
AS flights2 LEFT JOIN (SELECT year, month, day, AVG(temp) 
AS atemp, AVG(humid) AS ahumid FROM weather WHERE origin='EWR'
 GROUP BY year, month, day) AS weather2 ON flights2.year=weather2.year 
 AND flights2.month=weather2.month AND flights2.day=weather2.day""", conn)

task17_sql.head()
```

```python
# Filter flights data to include only records where the 'origin' is 'EWR'.
flights17 = flights[flights['origin']=='EWR']

# Filter weather data to include only records where the 'origin' is 'EWR'.
# Group the filtered weather data by 'year', 'month', and 'day', calculating the mean of 'temp' and 'humid'.
# Reset the index of the grouped weather data.
weather17 = (weather[weather['origin']=='EWR'].groupby(['year','month','day']).agg(atemp=('temp','mean'),
                                                                                   ahumid=('humid','mean')).reset_index())

# Merge the filtered flights data with the grouped weather data based on 'year', 'month', and 'day' using a left join.
task17_my = (pd.merge(flights17,weather17,left_on=['year','month','day'],right_on=['year','month','day'], how='left'))
task17_my.head(5)
```
